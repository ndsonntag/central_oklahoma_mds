```python
import numpy as np
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

#Plot basemap
fig = plt.figure(figsize = (6,4))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.LambertConformal(central_longitude=-97.5085, central_latitude=35.4819))
ax.set_extent([-102, -93, 33, 38], crs=ccrs.PlateCarree())

# Add cartopy features
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.STATES)

# Add features to show location of Oklahoma City and area that environmental parameters are averaged over
ax.add_patch(mpatches.Circle(xy=[-97.5085, 35.4819], radius=0.61, color='#301934', alpha=0.3, transform=ccrs.PlateCarree()))
ax.add_patch(mpatches.Circle(xy=[-97.5085, 35.4819], radius=0.05, color='black', transform=ccrs.PlateCarree()))


fig.tight_layout()
#plt.savefig('figure_01.jpg')
```
