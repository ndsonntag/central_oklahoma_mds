```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in data file using pandas
df = pd.read_csv("central_ok_mds_env_FINAL.csv")

# Create baseplot
plt.rcParams['figure.figsize'] = 15, 3

# Set custom color palette
palette = ['#ADD8E6', '#FF0000']
sns.set_palette(palette)

# Plot year data divided by label
ax = plt.subplot(1, 4, 1)
ax.set_title('(a) Number of Products per Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Products')
bins = np.arange(min(df.year), max(df.year + 1) + 1, 1)
sns.histplot(data = df, x = 'year', hue = 'label', bins = bins, legend = False)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.xticks(bin_centers, ['2014', '', '2016', '', '2018', '', '2020', '', '2022', '', '2024'])

# Plot month data divided by label
ax = plt.subplot(1, 4, 2)
ax.set_title('(b) Number of Products by Month')
ax.set_xlabel('Month')
ax.set_ylabel('Number of Products')
bins = np.arange(min(df.month), max(df.month + 1) + 1, 1)
sns.histplot(data = df, x = 'month', hue = 'label',  bins = 12, legend = False)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.xticks(bin_centers, ['Jan', '', 'Mar', '', 'May', '', 'Jul', '', 'Sep', '', 'Nov', ''])

# Plot day data divided by label
ax = plt.subplot(1, 4, 3)
ax.set_title('(c) Number of Products by Day')
ax.set_xlabel('Day')
ax.set_ylabel('Number of Products')
bins = np.arange(min(df.day), max(df.day + 1) + 1, 1)
sns.histplot(data = df, x = 'day', hue = 'label',  bins = 31, legend = False)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.xticks(bin_centers, ['', '', '', '', '5', '', '', '', '', '10', '', '', '', '', '15',
                        '', '', '', '', '20','', '', '', '', '25', '', '', '', '', '30', ''])
# Plot time data divided by label
ax = plt.subplot(1, 4, 4)
ax.set_title('(d) Number of Products by Hour')
ax.set_xlabel('Hour (UTC)')
ax.set_ylabel('Number of Products')
bins = np.arange(min(df.time), max(df.time + 1) + 1, 1)
sns.histplot(data = df, x = 'time', hue = 'label',  bins = 24, legend = False)
bin_centers = (bins[:-1] + bins[1:]) / 2
plt.xticks(bin_centers, ['0','', '', '', '', '5', '', '', '', '', '10', '', '', '', '', 
                         '15','', '', '', '', '20','', '', ''])
ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1), labels=['Watch', 'MD'])

plt.tight_layout()
plt.savefig('figure_6.jpg')
```
